```python
url = 'https://en.wikipedia.org/wiki/Peter_Pan_(1953_film)'
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Loading a webpage
page_req = requests.get(url)
soup = BeautifulSoup(page_req.content)
```

```python
info_box = soup.find(class_="infobox vevent")
#info_box.info_box = soup.find(class_="infobox vevent")
info_rows  = info_box.find_all("tr")
for i in info_rows:
    print(i.prettify())
```

```python
movie_info  = {}

def get_convert_list(td):
    if td.find("li"):
        return [li.get_text(strip=True).replace("\xa0"," ") for li in td.find_all("li")]
    else:
        return td.find('td').get_text(strip=True).replace("\xa0"," ")

for index,row in enumerate(info_rows):
    if index == 0:
        movie_info['Title'] = row.find('th').get_text(strip=True)
    elif index == 1:
        poster_url = url+"#/media/"+row.find('a').get("href").split("/")[-1]
        movie_info['Poster URL'] = poster_url
    else:
        key =  row.find('th').get_text(strip=True)
        value = get_convert_list(row)
        movie_info[key] = value
        
```
